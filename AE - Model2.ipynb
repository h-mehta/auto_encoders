```python
batch = 64
num_epochs = 20
```

```python
def layer(inputs, num, drop= False):
        
    lrelu = LeakyReLU()(inputs)

    conv = Conv2D(num, (3,3), strides=(2,2), padding="same")(lrelu)    
        
    batch = keras.layers.BatchNormalization()(conv)
    op = batch
        
    if(drop):
        drop = Dropout(0.5)(batch)
        op = drop
        
    return op
```

```python
def base():
    inputs = keras.layers.Input(shape=(128,128,1)) #128x128
    
    ip = Conv2D(4, (3,3), strides=(1,1), padding="same")(inputs) #128x128x4
    
    conv1 = layer(ip, 32) #64x64x32
    
    conv2 = layer(conv1,64) #32x32x64
 
    deconv1 = Conv2DTranspose(32, (3, 3), strides =(2,2), padding="same", activation="elu")(conv2) #64x64x32
    batch1 = keras.layers.BatchNormalization()(deconv1)
    batch1 = ReLU()(batch1)
    drop1 = Dropout(0.5)(batch1)
    
    deconv2 = Conv2DTranspose(16, (3, 3), strides =(2,2), padding="same", activation="elu")(drop1) #128x128x16
    batch2 = keras.layers.BatchNormalization()(deconv2)
    batch2 = ReLU()(batch2)
    drop2 = Dropout(0.5)(batch2)
    
    conv3 =  Conv2D(4, (3,3), strides=(1,1), padding="same")(drop2) #128x128x4
    
    op = Conv2D(1, (3, 3), strides =(1,1), padding="same", activation="sigmoid")(conv3)

    return Model(inputs,op)
```

```python
G = base()
G.compile(loss="logcosh", optimizer= 'adamax' , metrics="mse" )
G.summary()
```

```python
def lr_sch(epoch):
    if epoch < 2:
        return 1e-2
    if epoch < 4:
        return 1e-3
    if epoch < 6:
        return 1e-5
    if epoch < 10:
        return 1e-7
    if epoch < 25:
        return 1e-9
```

```python
fname = './output_file_2.h5'
```

```python
lr_scheduler = keras.callbacks.LearningRateScheduler(lr_sch)
lr_reducer = keras.callbacks.ReduceLROnPlateau(monitor='loss', factor=0.5, patience=3, mode='min')
checkpointer = keras.callbacks.ModelCheckpoint(filepath=fname, verbose=1, monitor='loss', mode='min',
                                               save_weights_only=True, save_best_only=True)
```

```python
class My_Generator(keras.utils.all_utils.Sequence):
#class My_Generator(keras.utils.Sequence):

    def __init__(self, image_filenames, groundtruth_filenames, batch_size= batch):
        self.image_filenames, self.groundtruth_filenames = image_filenames, groundtruth_filenames
        self.batch_size = batch_size
        
    def __len__(self):
        return np.uint16(np.ceil(len(self.image_filenames) / float(self.batch_size)))

    def __getitem__(self, idx):
        batch_x = self.image_filenames[idx * self.batch_size:(idx + 1) * self.batch_size]
        batch_y = self.groundtruth_filenames[idx * self.batch_size:(idx + 1) * self.batch_size]

        ip = np.array([(cv2.imread(file_name,0)/255) for file_name in batch_x],dtype=np.float32)

        op = np.array([((cv2.imread(file_name,0))/255) for file_name in batch_y],dtype=np.float32)
        
        return (ip, op)   
```

```python
class My_val_Generator(keras.utils.all_utils.Sequence):
#class My_val_Generator(keras.utils.Sequence):

    def __init__(self, image_filenames, groundtruth_filenames, batch_size= batch):
        self.image_filenames, self.groundtruth_filenames = image_filenames, groundtruth_filenames
        self.batch_size = batch_size
        
    def __len__(self):
        return np.uint16(np.ceil(len(self.image_filenames) / float(self.batch_size)))

    def __getitem__(self, idx):
        batch_x = self.image_filenames[idx * self.batch_size:(idx + 1) * self.batch_size]
        batch_y = self.groundtruth_filenames[idx * self.batch_size:(idx + 1) * self.batch_size]
        
        return (np.array([(cv2.imread(file_name,0)/255) for file_name in batch_x],dtype=np.float32), np.array([((cv2.imread(file_name,0))/255) for file_name in batch_y],dtype=np.float32))
```

```python
my_training_batch_generator = My_Generator(X_train, y_train, batch_size= batch)
my_validation_batch_generator = My_val_Generator(X_val, y_val, batch_size= batch)
```

```python
hist = G.fit(my_training_batch_generator, steps_per_epoch=(len(X_train) // batch),                                          
                                          epochs=num_epochs,
                                          callbacks=[lr_reducer, lr_scheduler, checkpointer],
                                          verbose=1,
                                          validation_data=my_validation_batch_generator,
                                          validation_steps=(len(X_val) // batch),
                                          workers=100)

G.save_weights(fname, overwrite=True)
```
